14. What is variance, and how is it calculated?

A:  Variance is a statistical measure that quantifies the spread or dispersion of a set of numbers.
    
    Find the average of all the numbers in your dataset.
    Subtract the mean from each individual number in the dataset.
    Square each of the differences you calculated in the previous step.
    Divide the sum of all the squared differences by the number of data points (for a population) or the number of data points minus 1 (for a sample).

```python
# 4. Implement sampling techniques like random sampling and stratified sampling

def stratified_sampling(df, stratum_column, sample_size_per_stratum):
    if sample_size_per_stratum > df[stratum_column].value_counts().min():
        try:

            raise ValueError("Sample size per stratum cannot be larger than the smallest stratum size.")
        
        except ValueError as vE:

            print(vE)

    return df.groupby(stratum_column, group_keys=False).apply(lambda x: x.sample(sample_size_per_stratum))

data = {'ID': range(1, 101),
    'Group': ['A'] * 40 + ['B'] * 35 + ['C'] * 25,
    'Value': [random.randint(1, 100) for _ in range(100)]}
df = pd.DataFrame(data)

stratum_column = 'Group'
sample_size_per_stratum = 5
stratified_sample = stratified_sampling(df, stratum_column, sample_size_per_stratum)
print(f"Stratified sample:\n{stratified_sample}")


def random_sampling(population, sample_size):
    if sample_size > len(population):
        try:
            raise ValueError("Sample size cannot be larger than the population size.")
        except ValueError as vE:
            print(vE)
    return random.sample(population, sample_size)

population = list(range(1, 101))
sample_size = 20
random_sample = random_sampling(population, sample_size)
print(f"Random sample of size {sample_size}: {random_sample}")
```

```python
#  9. Write a Python script to calculate covariance between two datasets
def calculate_covariance(dataset1, dataset2):

    if not dataset1 or not dataset2 or len(dataset1) != len(dataset2):
        return None

    array1 = np.array(dataset1)
    array2 = np.array(dataset2)

    mean1 = np.mean(array1)
    mean2 = np.mean(array2)

    covariance = np.sum((array1 - mean1) * (array2 - mean2)) / (len(array1) - 1)
    return covariance

dataset_x = [2, 4, 6, 8]
dataset_y = [1, 3, 5, 7]

covariance_xy = calculate_covariance(dataset_x, dataset_y)
print(f"Covariance between dataset_x and dataset_y: {covariance_xy}")

covariance_matrix = np.cov(dataset_x, dataset_y)
print(f"Covariance matrix:\n{covariance_matrix}")
```

```python
#  10. Write a Python script to calculate the correlation coefficient between two datasets.


def calculate_correlation_coefficient(data_x, data_y):
    try:
        correlation_coefficient = np.corrcoef(data_x, data_y)[0, 1]
        return correlation_coefficient
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

dataset_x = [1, 2, 3, 4, 5]
dataset_y = [2, 4, 1, 3, 5]

correlation = calculate_correlation_coefficient(dataset_x, dataset_y)

if correlation is not None:
    print(f"The Pearson correlation coefficient is: {correlation}")
```

```python
#  12. Implement and compare simple random sampling and systematic sampling.

def simple_random_sampling(population, n):
    if n > len(population):
        return "Sample size exceeds population size."
    return random.sample(population, n)

population = list(range(1, 101))
n = 10
simple_random_sample = simple_random_sampling(population, n)
print(f"Simple Random Sample: {simple_random_sample}")


def systematic_sampling(population, n):
    if n > len(population):
         return "Sample size exceeds population size."
    
    k = len(population) // n 
    start = random.randint(0, k - 1)  
    indices = range(start, len(population), k)
    return [population[i] for i in indices]

population = list(range(1, 101))
n = 10
systematic_sample = systematic_sampling(population, n)
print(f"Systematic Sample: {systematic_sample}")
```

```python
# 13. Calculate the mean, median, and mode of grouped data

def calculate_mean_grouped(grouped_data):
    total_frequency = 0
    sum_midpoint_frequency = 0

    for lower_bound, upper_bound, frequency in grouped_data:
        midpoint = (lower_bound + upper_bound) / 2
        sum_midpoint_frequency += midpoint * frequency
        total_frequency += frequency

    if total_frequency == 0:
        return 0  
    else:
        return sum_midpoint_frequency / total_frequency

grouped_data = [
    (0, 9, 5),
    (10, 19, 10),
    (20, 29, 15),
    (30, 39, 8)
]
mean = calculate_mean_grouped(grouped_data)
print(f"Estimated Mean: {mean}")
```

```python
#  19.  Compare two datasets using their standard deviations.

def compare_std_dev(dataset1, dataset2):
    std_dev1 = np.std(dataset1)
    std_dev2 = np.std(dataset2)

    if std_dev1 < std_dev2:
        return f"Dataset 1 has a smaller standard deviation ({std_dev1:.2f}) than Dataset 2 ({std_dev2:.2f}). Data points in Dataset 1 are more clustered around the mean."
    elif std_dev1 > std_dev2:
        return f"Dataset 1 has a larger standard deviation ({std_dev1:.2f}) than Dataset 2 ({std_dev2:.2f}). Data points in Dataset 1 are more spread out than in Dataset 2."
    else:
        return f"Dataset 1 and Dataset 2 have the same standard deviation ({std_dev1:.2f})."

data1 = [10, 12, 15, 18, 20]
data2 = [5, 10, 15, 20, 25]

comparison_result = compare_std_dev(data1, data2)
print(comparison_result)
```

```python
# 22. Generate a dataset and implement both variance and standard deviation computations

dataset = [random.randint(1, 100) for _ in range(10)]
print("Dataset:", dataset)


def calculate_mean(data):
    return sum(data) / len(data)

mean = calculate_mean(dataset)
print("Mean:", mean)


def calculate_variance(data, mean):
    squared_differences = [(x - mean) ** 2 for x in data]
    return sum(squared_differences) / len(data)

variance = calculate_variance(dataset, mean)
print("Variance:", variance)


def calculate_standard_deviation(data, mean):
    return math.sqrt(calculate_variance(data, mean))

standard_deviation = calculate_standard_deviation(dataset, mean)
print("Standard Deviation:", standard_deviation)
```

```python
#  23. Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn.

np.random.seed(42)
data_normal = np.random.normal(0, 1, 1000)
data_skewed_pos = np.random.exponential(1, 1000)
data_skewed_neg = -np.random.exponential(1, 1000)
data_kurtosis_high = np.concatenate([np.random.normal(-3, 0.5, 250), np.random.normal(3, 0.5, 250), np.random.normal(0, 1, 500)])

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.histplot(data_normal, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title(f'Normal Distribution (Skewness: {skew(data_normal):.2f}, Kurtosis: {kurtosis(data_normal):.2f})')

sns.histplot(data_skewed_pos, kde=True, ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title(f'Positively Skewed Distribution (Skewness: {skew(data_skewed_pos):.2f}, Kurtosis: {kurtosis(data_skewed_pos):.2f})')

sns.histplot(data_skewed_neg, kde=True, ax=axes[1, 0], color='salmon')
axes[1, 0].set_title(f'Negatively Skewed Distribution (Skewness: {skew(data_skewed_neg):.2f}, Kurtosis: {kurtosis(data_skewed_neg):.2f})')

sns.histplot(data_kurtosis_high, kde=True, ax=axes[1, 1], color='gold')
axes[1, 1].set_title(f'High Kurtosis Distribution (Skewness: {skew(data_kurtosis_high):.2f}, Kurtosis: {kurtosis(data_kurtosis_high):.2f})')

plt.tight_layout()
plt.show()
```
